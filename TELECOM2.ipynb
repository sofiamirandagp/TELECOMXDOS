```python
variables_categoricas = [
    'genero', 'tipo_contrato', 'metodo_pago',
    'servicio_tv', 'servicio_peliculas',
    'seguridad_en_linea', 'soporte_tecnico'
]
```

```python
def analizar_cancelaciones(df, variable):
    df_filtrado = df[df[variable] != 'No internet service']  # Filtra "No internet service" si existe
    df_filtrado = df_filtrado.copy()

    # Agrupar y calcular proporciones
    tabla = pd.crosstab(df_filtrado[variable], df_filtrado['cancelacion'], normalize='index') * 100
    tabla.columns = ['Permanecen (%)', 'Cancelan (%)']

    print(f'\nAnÃ¡lisis de la variable: {variable}')
    print(tabla.round(2))

    # GrÃ¡fico
    tabla_plot = tabla.reset_index().melt(id_vars=variable, value_name='Porcentaje', var_name='Estado')
    plt.figure(figsize=(7, 4))
    sns.barplot(data=tabla_plot, x=variable, y='Porcentaje', hue='Estado', palette=['#4CAF50', '#F44336'])
    plt.title(f'DistribuciÃ³n de cancelaciones por {variable}')
    plt.ylabel('Porcentaje')
    plt.xticks(rotation=30)
    plt.legend(title='Estado')
    plt.tight_layout()
    plt.show()

# Aplicar anÃ¡lisis a todas las variables
for var in variables_categoricas:
    analizar_cancelaciones(datos, var)
```

```python

# Variables numÃ©ricas a analizar
variables_numericas = ['cargo_total', 'meses_contrato']

for var in variables_numericas:
    print(f"AnÃ¡lisis de la variable: {var}")
    resumen = datos.groupby('cancelacion')[var].describe()
    print(resumen, "\n")

    # Boxplot comparando la distribuciÃ³n segÃºn cancelaciÃ³n
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='cancelacion', y=var, data=datos, hue='cancelacion',
                palette=['#4CAF50', '#F44336'], dodge=False)
    plt.title(f'DistribuciÃ³n de {var} segÃºn cancelaciÃ³n')
    plt.xlabel('CancelaciÃ³n (False=No, True=SÃ­)')
    plt.ylabel(var)
    plt.legend(title='CancelaciÃ³n', loc='upper right')
    plt.show()

    # Histograma con KDE para cada grupo
    plt.figure(figsize=(8, 5))
    sns.histplot(data=datos, x=var, hue='cancelacion', kde=True,
                 palette=['#4CAF50', '#F44336'], alpha=0.5)
    plt.title(f'Histograma y KDE de {var} por estado de cancelaciÃ³n')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.show()
```

**Modelo 1: RegresiÃ³n LogÃ­stica (con normalizaciÃ³n)**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix

# Reemplazar 'Yes'/'No' por 1/0 en todas las columnas que tengan esos valores
for col in datos.columns:
    if datos[col].dtype == 'object':
        datos[col] = datos[col].replace({'Yes': 1, 'No': 0})


# Convertir variables categÃ³ricas a dummies (one-hot encoding)
columnas_categoricas = ['genero', 'lineas_multiples', 'seguridad_en_linea', 'respaldo_en_linea',
                       'proteccion_dispositivos', 'soporte_tecnico', 'servicio_tv', 'servicio_peliculas',
                       'tipo_internet', 'tipo_contrato', 'metodo_pago']

datos_dummies = pd.get_dummies(datos, columns=columnas_categoricas, drop_first=True)

# Convertir la variable'cancelacion' a numÃ©rica (True=1, False=0)
datos_dummies['cancelacion_num'] = datos_dummies['cancelacion'].astype(int)

# Eliminar columnas irrelevantes para el modelo
if 'ID_cliente' in datos_dummies.columns:
    datos_dummies = datos_dummies.drop(columns=['ID_cliente'])
if 'cancelacion' in datos_dummies.columns:
    datos_dummies = datos_dummies.drop(columns=['cancelacion'])

# Eliminar filas con NaNs que puedan haber quedado
datos_dummies = datos_dummies.dropna()

# Definir variables predictoras y objetivo
X = datos_dummies.drop(columns=['cancelacion_num'])
y = datos_dummies['cancelacion_num']

# Dividir en conjunto de entrenamiento y prueba (70% / 30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Crear pipeline con escalador y regresiÃ³n logÃ­stica
pipeline_logistica = Pipeline([
    ('scaler', StandardScaler()),
    ('logistic_regression', LogisticRegression(max_iter=1000, random_state=42))
])

# Entrenar el modelo
pipeline_logistica.fit(X_train, y_train)

# Predecir y evaluar
y_pred = pipeline_logistica.predict(X_test)

print("Reporte de clasificaciÃ³n:\n", classification_report(y_test, y_pred))
print("Matriz de confusiÃ³n:\n", confusion_matrix(y_test, y_pred))
```

**JustificaciÃ³n de la normalizaciÃ³n:**

En este modelo aplicamos una etapa de normalizaciÃ³n a las variables numÃ©ricas mediante *StandardScaler*, que transforma los datos para que tengan una media cercana a cero y una desviaciÃ³n estÃ¡ndar igual a uno.

La normalizaciÃ³n evita que las variables con valores numÃ©ricos mayores tengan un peso desproporcionado durante el entrenamiento. AsÃ­, todas las variables quedan en una escala comparable, lo que previene que el modelo se sesgue hacia aquellas con magnitudes mÃ¡s altas.

Esto es especialmente importante en modelos como la regresiÃ³n logÃ­stica, que realizan cÃ¡lculos sensibles a la escala de los datos. Al normalizar, garantizamos que el modelo aprenda de manera equilibrada de todas las variables, mejorando su capacidad predictiva.
